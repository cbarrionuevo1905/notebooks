```python
#Setea el estilo de los graficos

# algún bug cuando lo hice:
plt.figure(1)
plt.close(1)

import matplotlib as mpl

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 11

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})

#%%  Fin - Inicialización de librerías
```

```python
# Librerías externas NumPy, SciPy y Matplotlib
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import TransferFunction
from scipy import signal as sig

# Librería de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
```

```python
w0 = 1

num = np.array([1, -w0])
den = np.array([1, w0])

my_tf = TransferFunction(num, den)

_, axes_hdl = bodePlot(my_tf, fig_id=1)

plt.sca(axes_hdl[0])
plt.ylim([-1,1])

pzmap(my_tf, fig_id=2) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3)
```
