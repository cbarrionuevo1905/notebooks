# EJERCICIO 4: Transferencia normalizada

```python
# Librerías externas NumPy, SciPy y Matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from scipy.signal import TransferFunction
from scipy import signal as sig

# Librería de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

w0 = 1/0.509

num = np.array([w0])
den = np.array([1, 2.5,3.14,w0])

my_tf = TransferFunction(num, den)

_, axes_hdl = bodePlot(my_tf, fig_id=1)

plt.sca(axes_hdl[0])
plt.ylim([-70,10])


pzmap(my_tf, fig_id=2) #S plane pole/zero plot

GroupDelay(my_tf, fig_id=3)
```

Como podemos visualizar, no, no es un filtro Butter ya que $ε \neq 1$ , sin embargo, puedo determinar un factor de normalizacion $w_b$, que en su expresion contenga el ripple, tal que pueda visualizar al filtro como un Butter.
